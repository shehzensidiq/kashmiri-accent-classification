# MFCC csv file making

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow

def getImageFeatures(folder,i):
    image = None
    for (directoryPath, directoryName, filenames ) in os.walk(folder):
        for filename in filenames:
            if filename == i:
                image = imread(directoryPath+"/"+filename)
                # print(image.shape)
                # print(i, filename, "Extracting Features")
            # else:
            #     print("False")
    return image
            # print(filename)
            # imshow(image)
```

```python
imageLabels=[]
from os import path
def imageCsv(folderName, csvName):
    label = []
    for (directoryPath, directoryName, filenames ) in os.walk(folderName+csvName):
        if directoryName != 'processed':
            for name in filenames:
                labelDistrict = name.split("_")[0]
                imageLabels.append([name, labelDistrict])
                # testLabel = [filename, labelDistrict]
                imageFeature = getImageFeatures(folderName+csvName, name)
                with open(f'{BASE_PATH}/{csvName}.csv', 'a', newline="") as file:
                    writer = csv.writer(file)
                    writer.writerow([name, imageFeature, labelDistrict])
                if path.exists(folderName+csvName+"/"+name):
                    shutil.move(folderName+csvName+"/"+name, f'{folderName+csvName}/processed/'+name)
                else:
                    # print(f"Path does not exist: {folderName+csvName}/{name}")
                    pass
                # os.remove(folderName+csvName+"/"+filename)

    print(imageLabels)
```

```python
imageCsv(f'{BASE_PATH}', 'mfcc')
```
