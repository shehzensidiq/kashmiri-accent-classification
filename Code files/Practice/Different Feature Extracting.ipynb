# Amplitude Envelop extraction

```python
import pandas as pd
import numpy as np
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
filePath = "/content/drive/MyDrive/Masters Project/Dataset/final audio/14c281fO1r22N51i80a7P82520.wav"
ipd.Audio(filePath)
```

```python
audioData, sampleRate = librosa.load(filePath)
import librosa.display
plt.figure(figsize=(10,5))
plt.ylim(-1,1)
librosa.display.waveplot(audioData, sampleRate)
```

```python
mfcc = librosa.feature.mfcc(audioData, sampleRate, n_mfcc=12)
```

```python
mfcc.shape
```

```python
print(mfcc)
```

```python
plt.figure(figsize=(15,10))
librosa.display.specshow(mfcc, sr=sampleRate, x_axis='time', y_axis='log')
```

```python
plt.figure(figsize=(15,10))
librosa.display.specshow(mfcc, sr=sampleRate, x_axis='time', y_axis='log')
plt.axis('off')
plt.savefig('new.png')
```
