```python
import os, sys
import argparse
from pydub import AudioSegment
# formats = [".m4a"]
formats = ['.m4a', '.mp3']
```

```python
def convertAudioToWav(folderName):
    # convert all other formats to m4a - easy as compared to giving each extension
    # for filename in os.listdir(folderName):
    #    initialFile = os.path.join(folderName,filename)
    #    if not os.path.isfile(initialFile): continue
    #    oldExt = os.path.splitext(filename) # splits the name on extension
    #    newExt = initialFile.replace('.tmp', '.m4a')
    #    outputFile = os.rename(initialFile, newExt)
    #    print("")

    # convert m4a to wav format
    for (dirpath, dirnames, filenames) in os.walk(folderName):
        for filename in filenames:
            if filename.endswith(tuple(formats)):
                filepath = dirpath + '/' + filename
                (path, fileExt) = os.path.splitext(filepath)
                fileExt = fileExt.replace('.', '')
                try:
                    audioFile = AudioSegment.from_file(filepath, fileExt)
                    wav = filename.replace(fileExt, 'wav')
                    wavPath = dirpath + '/' + wav
                    print("Cnverting..." + str(filepath))
                    finalFile = audioFile.export(wavPath, format='wav')
                    os.remove(filepath)
                    print("Conversion complete")
                except:
                    print("Error while converting...")
```

```python
path = '/content/drive/MyDrive/Masters Project/Dataset/Audio'
convertAudioToWav(path)
```
