```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
```

```python
BASE_PATH = '/content/drive/MyDrive/Masters Project/Dataset/'
```

```python
imageData= []
for (dirPath, dirName, filenames) in os.walk(BASE_PATH+"mfcc/"):
    for filename in filenames:
        image = tf.keras.preprocessing.image.load_img(BASE_PATH+"mfcc/"+filename)
        print(image.size)
        imageArr = tf.keras.preprocessing.image.img_to_array(image)
        imageData.append([imageArr, filename.split("_")[0]])
```

```python
labels =[]
images = []
for data in imageData:
    labels.append(data[1])
    images.append(data[0])
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Conv2D(50, (5,5), activation='relu', input_shape=inputShape))
```

```python
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(4))
```

```python
model.compile(optimizer="adam", loss='categorical_crossentropy',
              metrics=["accuracy"])
```
