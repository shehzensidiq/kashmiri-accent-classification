```python
def getImageArray(dirpath, filename):
    image = keras.preprocessing.image.load_img(os.path.join(dirpath,filename))
    image = keras.preprocessing.image.img_to_array(image)
    image = image.tolist()
    return image
```

```python
mfcc24 = f"{BASE_PATH}mfcc/mfcc24"
mfcc24data = {
    "labels": [],
    "imageArray": []
}
for (dirpath, dirname, filenames) in os.walk(mfcc24):
    for filename in filenames:
        print(f"Processing file: {filename}")
        mfcc24data["labels"].append(filename.split("_")[0])
        image = getImageArray(dirpath, filename)
        mfcc24data["imageArray"].append(image)
with open(f"{BASE_PATH}mfcc24.json", 'w') as file:
    json.dump(mfcc24data, file, indent=4)
```

```python
# loading back the image json files
import numpy as np
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
from sklearn.model_selection import train_test_split
def getData():
    data=None
    with open(os.path.join(BASE_PATH, 'mfcc24.json'), 'r') as fp:
        data = json.load(fp)
    X = np.array(data["imageArray"])
    y = np.array(data["labels"])
    xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.1,random_state=112)
    # ytrain = tf.keras.utils.to_categorical(ytrain)
    # ytest = tf.keras.utils.to_categorical(ytest)
    ytrain = le.fit_transform(ytrain)
    ytest = le.transform(ytest)
    return xtrain, ytrain, xtest, ytest
    # print(X[0].shape)
    # print(X.shape)
```

```python
xtrain, ytrain, xtest, ytest = getData()
```

```python
def genModel(lr=0.001):
    model = keras.models.Sequential()
    # first layer
    model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation="relu", input_shape=inputShape,
                                  kernel_regularizer=keras.regularizers.l2(0.001)))
    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.MaxPool2D((3,3), strides=(2,2), padding="same"))

    # 2 con layer
    model.add(keras.layers.Conv2D(32, kernel_size=(3,3), activation="relu",
                                  kernel_regularizer=keras.regularizers.l2(0.001)))
    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.MaxPool2D((3,3), strides=(2,2), padding="same"))

    # 3 conv 
    model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation="relu",
                                  kernel_regularizer=keras.regularizers.l2(0.001)))
    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.MaxPool2D((3,3), strides=(2,2), padding="same"))

    #flatten layer
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(64, activation="relu"))
    model.add(keras.layers.Dropout(0.1))

    model.add(keras.layers.Dense(4, activation="softmax"))
    ## compile
    optimizer = keras.optimizers.Adam(lr)
    model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    print(model.summary())
    return model
```

```python
def trainModel():
    lr = 0.001
    model = genModel(lr)
    model.fit(xtrain, ytrain, batch_size=32, validation_data=(xtest, ytest), epochs=40)
    testError, testAcc = model.evaluate(xtest, ytest)
    print(f"Test error: {testError}, Test acc: {testAcc}")
    model.save("model1.h5")
```

```python
trainModel()
```

```python
def getSpecs():
    data=None
    le = LabelEncoder()
    with open(os.path.join(BASE_PATH, 'specs.json'), 'r') as fp:
        data = json.load(fp)
    X = np.array(data["imageArray"])
    y = np.array(data["labels"])
    xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.1,random_state=112)
    # ytrain = tf.keras.utils.to_categorical(ytrain)
    # ytest = tf.keras.utils.to_categorical(ytest)
    ytrain = le.fit_transform(ytrain)
    ytest = le.transform(ytest)
    return xtrain, ytrain, xtest, ytest
```

```python
def trainModel():
    lr = 0.001
    model = genModel(lr)
    hist = model.fit(xtrain, ytrain, batch_size=32, epochs=3)
    testError, testAcc = model.evaluate(xtest, ytest)
    print(f"Test error: {testError}, Test acc: {testAcc}")

    import matplotlib.pyplot as plt
    plt.plot(hist.history['accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(hist.history['loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    model.save("modelSpec.h5")
```
