```python
# stretch the time
def stretch(data, speed):
    stretched = lb.effects.time_stretch(data, speed)
    name = row[1].split("_")[0]
    currentStamp = datetime.datetime.now().strftime(dateFormat)
    sf.write(f'{BASE_PATH}Audio/Augmented/{name}_{currentStamp}.wav', stretched,
             sampleRate, format="wav", subtype="PCM_24")
    # with open(f'{BASE_PATH}audio_meta.csv', 'a', newline="") as file:
    #     writer = csv.writer(file)
    #     writer.writerow([name+"_"+currentStamp+".wav", name])
    print(f"Speeding the {BASE_PATH}Audio/Final Audio/{name}_{currentStamp}.wav")
```

```python
for index, row in metadata.iterrows():
    filename = f"{BASE_PATH}/Audio/Final Data/{row[0]}"
    data, sampleRate = lb.load(filename)
    shiftPitch(data, sampleRate, 7)
    shiftPitch(data, sampleRate, 5)
    stretch(data, 1.3)
    stretch(data, 0.85)
```
